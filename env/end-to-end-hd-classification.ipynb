### Modelling

```python
df.head()
```

```python
#Split data into X and Y

X = df.drop("target",axis=1)
Y = df["target"]
```

```python
#Split data into train and test sets

np.random.seed(42)

# Split into train and test set

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
X_train
```

```python
Y_train, len(Y_train)
```

# We are going to try 3 ml models
1. Logistin Regression
2. K-nearest Neighbours classifier
3. Random Forest Classifier

```python
#Put  models in a dictionary
models = {"Logistic Regression" : LogisticRegression(),
          "KNN" : KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier()}
# Create a function to fit and socre models

def fit_and_score(models,X_train,X_test,Y_train,Y_test):
    """
    Fits and evaluates given ML models
    """
    np.random.seed(42)

    model_scores={}
    #Loop through models

    for name, model in models.items():
        model.fit(X_train,Y_train)
        model_scores[name] = model.score(X_test,Y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models,X_train,X_test,Y_train,Y_test)

model_scores
```

### Model Comparison

```python
model_compare = pd.DataFrame(model_scores,index=["accuracy"])
model_compare.T.plot.bar();
plt.show();
```

### Hyperparameter Tuning

```python
train_scores = []
test_scores = []

#Create a list of different values for n_neighbours 
neighbours = range(1,21)

knn = KNeighborsClassifier()

for i in neighbours:
    knn.set_params(n_neighbors = i)

    knn.fit(X_train,Y_train)

    train_scores.append(knn.score(X_train,Y_train))

    test_scores.append(knn.score(X_test,Y_test))

train_scores
```

```python
test_scores
```

```python
plt.plot(neighbours, train_scores,label="train score")
plt.plot(neighbours, test_scores,label="test score")
plt.xlabel("Number of neighbours")
plt.ylabel("Model score")
plt.legend()
plt.show();

```

## Hyperparameter tuning with RandomizedSearchCV

```python
#Create a hyperparameter grid for LogisticRegression
log_reg_grid = {"C" : np.logspace(-4,4,20),
               "solver" : ["liblinear"]}
# Create a hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators" : np.arange(10,1000,50),
           "max_depth" : [None,3,5,10],
           "min_samples_split" : np.arange(2,20,2),
           "min_samples_leaf": np.arange(1,20,22)}
```

```python
np.random.seed(42)

#Setup random hyperparamteer search for LR

rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                               param_distributions=log_reg_grid,
                               cv=5,
                               verbose=True,
                               n_iter=20)
#Fit random hyperparameter search model for LR
rs_log_reg.fit(X_train,Y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(X_test,Y_test)
```

```python
np.random.seed(42)

rs_rf = RandomizedSearchCV(RandomForestClassifier(),
                           param_distributions=rf_grid,
                           cv =5,
                           n_iter = 20,
                           verbose = True)
# Fit random hyperparamter search model for RFC
rs_rf.fit(X_train,Y_train)
```

```python
#Find the best hyperparameters
rs_rf.best_params_
```

```python
#Evaluate the randomized search RandomForestClassifer model

rs_rf.score(X_test,Y_test)
```

## Hyperparamter Tuning with GridSeearchCV


```python
# Different hyperparameters for our LogisticRegression model
log_reg_grid = {"C" : np.logspace(-4,4,30),
               "solver" : ["liblinear"]}
gs_log_reg = GridSearchCV(LogisticRegression(),
                          param_grid=log_reg_grid,
                          cv=5,
                          verbose=2)
gs_log_reg.fit(X_train,Y_train)
```

```python
gs_log_reg.best_params_
```

```python
#Evaluate the grid search LR model
gs_log_reg.score(X_test,Y_test)
```

```python
sns.set(font_scale = 1.5)

def plot_conf_mat(Y_test,Y_preds):
    fig,ax = plt.subplots(figsize=(3,3))
    ax = sns.heatmap(confusion_matrix(Y_test,y_preds),
                    annot = True,
                    cbar = False)
plot_conf_mat(Y_test,y_preds)
plt.show();
```

```python
print(classification_report(Y_test,y_preds))
```

### calculate eval metrics using cross validations

```python
gs_log_reg.best_params_
```

```python
# Create a new classifier with best params
clf = LogisticRegression(C = 0.20433597178569418,
                         solver = "liblinear")
```

```python
cv_acc = cross_val_score(clf,X,Y,cv=5,scoring="accuracy")
cv_acc
```

```python
cv_acc = np.mean(cv_acc)
```

```python
cv_prec = cross_val_score(clf,X,Y,cv=5,scoring="precision")
cv_prec

cv_prec = np.mean(cv_acc)
cv_prec
```

```python
cv_recall = cross_val_score(clf,X,Y,cv=5,scoring="recall")
cv_recall = np.mean(cv_recall)

cv_recall
```

```python
cv_f1_score = cross_val_score(clf,X,Y,cv=5,scoring="f1")
cv_f1_score = np.mean(cv_f1_score)

cv_f1_score
```

```python
cv_metrics = pd.DataFrame({"Accuracy":cv_acc,
                           "Precision":cv_prec,
                           "Recall":cv_recall,
                           "F1":cv_f1_score},
                         index = [0])
cv_metrics.T.plot.bar(title = "Cross Validate Metrics")
plt.show();
```
